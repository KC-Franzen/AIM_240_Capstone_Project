```python
# List of library and resources needed
import pandas as pd
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
import string
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib

#nltk.download('stopwords')
#nltk.download('punkt')
```

```python
# Grabbing text file from path
energy_text = "C:/Users/krisf/Documents/AIM 240 Capstone Project/Energy_Text.txt"

# Open and read energy_text and join the document text to a string
with open(energy_text, 'r', encoding='utf-8') as file :
    read_text = ''.join(file.readlines())

pd.set_option('display.width',75)

# Tokenize the document into sentences
et_sentences = sent_tokenize(read_text)

# Print the sentences to verify
for i, sentence in enumerate(et_sentences) :
    print(f"Sentence {i+1}: {sentence}")
```

```python
# Setting the stopwords to English
stop_words = set(stopwords.words('english'))

# Function to clean sentences
def clean_sentence(sentence) :
    # Convert to lowercase
    sentence = sentence.lower()
    # Tokenize the sentence
    words = word_tokenize(sentence)
    # Remove stopwords and punctuation
    words = [word for word in words if word not in stop_words and word not in string.punctuation]
    return ' '.join(words)

# Clean the sentences
cleaned_et_sentences = [clean_sentence(sentence) for sentence in et_sentences]

# Print the sentences to verify
for i, sentence in enumerate(cleaned_et_sentences) :
    print(f"Sentence {i+1}: {sentence}")

print(type(cleaned_et_sentences))
```

```python
# Create DataFrame with sentences
clean_et_data = pd.DataFrame({'sentence': cleaned_et_sentences, 'label': 0})

# Save the DataFrame to a CSV file
clean_et_data.to_csv('clean_et_data_with_labels.csv', index=False)

# Load the saved CSV file
loaded_et_data = pd.read_csv('clean_et_data_with_labels.csv')

# Display the first few rows of the loaded DataFrame
print(loaded_et_data.head())
```
